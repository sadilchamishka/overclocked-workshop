```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import math
import random
```

## Matplotlib - Library

```python
data = pd.read_csv("data.csv")
data.head()
```

## Histograms

```python
plt.hist(data['SalePrice'], bins = 50)
plt.ylabel('SalePrice')
plt.xlabel('Id')
plt.suptitle("Sales Price Distribution")
plt.show()
```

### Plotting Multiple Histograms in an One Plot

```python
x = [random.gauss(3,1) for _ in range(400)]
y = [random.gauss(4,2) for _ in range(400)]

bins = np.linspace(-10, 10, 100)

plt.hist(x, bins, alpha=0.5, label='x')
plt.hist(y, bins, alpha=0.5, label='y')
plt.legend(loc='upper right')
plt.show()
```

## Line Graphs 

```python
year = [1960, 1970, 1980, 1990, 2000, 2010]
pop_pakistan = [44.91, 58.09, 78.07, 107.7, 138.5, 170.6]
plt.plot(year, pop_pakistan, color='g')
plt.xlabel('Countries')
plt.ylabel('Population in million')
plt.suptitle('Pakistan Population till 2010')
plt.show()
```

### Plotting Multiple Line Graphs in an One Plot

```python
x=[]
for i in range(-314,314):
   x.append(i/100)
ysin=[math.sin(i) for i in x]    #calculate Sin(x)
ycos=[math.cos(i) for i in x]    #calculate Cos(x)
plt.plot(x,ysin,label='sin(x)')  #specify label for the Sin(x) curve
plt.plot(x,ycos,label='cos(x)')  #specify label for the Cos(x) curve
plt.xticks([-3.14,-1.57,0,1.57,3.14],['-$\pi$','-$\pi$/2',0,'$\pi$/2','$\pi$'])
plt.legend()
plt.show()
```

## Subplots 

```python
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, sharex='col', sharey='row')
f.suptitle('Sharing x per column, y per row')
ax1.plot(x, y)
ax2.scatter(x, y)
ax3.scatter(x, 2 * y ** 2 - 1, color='r')
ax4.plot(x, 2 * y ** 2 - 1, color='r')
```

## Bar Chart

```python
objects = ('Python', 'C++', 'Java', 'Perl', 'Scala', 'Lisp')
y_pos = np.arange(len(objects))
performance = [10,8,6,4,2,1]

plt.bar(y_pos, performance, align='center', alpha=0.5)
plt.xticks(y_pos, objects)
plt.ylabel('Usage')
plt.title('Programming language usage')

plt.show()
```

## Pie Chart

```python
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.1, 0, 0, 0)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)

plt.axis('equal')
plt.show()
```

## Seaborn - Library

```python
import seaborn as sns
data = sns.load_dataset("tips")
data.head()
```

### Scatter Plot

```python
sns.relplot(x="total_bill", y="tip", 
            col="time", # Categorical variables that will determine the faceting of the grid.
            hue="smoker", # Grouping variable that will produce elements with different colors.
            style="smoker", # Grouping variable that will produce elements with different styles.
            size="size", # Grouping variable that will produce elements with different sizes.
            data=data)
```

### Plotting data while finding a linear model

```python
sns.lmplot(x="total_bill", y="tip", 
           col="time", # Categorical variables that will determine the faceting of the grid.
           hue="smoker", # Grouping variable that will produce elements with different colors.
           data=data)
```

### Visualizing Categorical variables

```python
sns.catplot(x="day", y="total_bill", 
            hue="smoker", # Grouping variable that will produce elements with different colors.
            kind="box", # Options are: "point", "bar", "strip", "swarm", "box", "violin", or "boxen"
            data= data)
```
