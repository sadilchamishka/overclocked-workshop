#### Add New Row

```python
data.reset_index(inplace=True)                         #reset the index of dataframe
random_row = np.random.randint(1, 6, data.shape[1])    #inserting new row
data.loc[-1] = random_row                              #add new row to last index
data.index = data.index + 1                            #increase the index of dataframe
data = data.sort_index()                               #sort dataframe by index
data
```

#### Merge Dataframes

```python
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],'B': ['B0', 'B1', 'B2', 'B3'],'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']})
df1
```

```python
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],'B': ['B4', 'B5', 'B6', 'B7'],'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']})
df2
```

##### Merge Vertically

```python
frames = [df1, df2]
merged_data = pd.concat(frames).reset_index(drop=True)
merged_data
# Can be easily merged with append
# df1.append(df2)
```

#### Merge Horizontally

```python
df2 = pd.DataFrame({'E': ['A4', 'A5', 'A6', 'A7'],'F': ['B4', 'B5', 'B6', 'B7'],'G': ['C4', 'C5', 'C6', 'C7'],
                    'H': ['D4', 'D5', 'D6', 'D7']})
```

```python
frames = [df1, df2]
merged_data = pd.concat(frames, axis=1).reset_index(drop=True)
merged_data
```

#### Join Opertation

```python
df2 = pd.DataFrame({'D': ['D2', 'D3', 'D6', 'D7'],
                  'F': ['F2', 'F3', 'F6', 'F7']},
                   )
df2
```

```python
pd.merge(df1, df2, on='D', how='inner')
```
